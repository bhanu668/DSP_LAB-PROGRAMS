```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,50,15])
plt.pie(y)
plt.show()
```

```python
y=np.array([35,50,15])
co=["red","blue","hotpink"]
plt.pie(y,colors=co,labels=["DSP","CD","WT"])
plt.show()
```

```python
y=np.array([35,50,15])
co=["red","blue","hotpink"]
plt.pie(y,colors=co,labels=["DSP","CD","WT"])
plt.legend(title="SUBJECTS")
plt.show()
```

```python
x=np.random.rand(40)
y=np.random.rand(40)
z=np.random.rand(40)
colors=np.random.rand(40)
plt.scatter(x,y,s=z*1000,c=colors)
plt.show()
```

```python
x=np.random.rand(40)
y=np.random.rand(40)
z=np.random.rand(40)
colors=np.random.rand(40)
plt.scatter(x,y,s=z*1000,c=colors,alpha=0.3)
plt.show()
```
